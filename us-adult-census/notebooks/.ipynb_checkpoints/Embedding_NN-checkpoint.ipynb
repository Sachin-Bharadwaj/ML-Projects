### create stratifiedfolds using train df

```python
### write create_folds function to output stratified folds for validation
def create_folds(df, num_folds=5, target_colname=None, save_df=False):
    # create kfold column
    df["kfolds"] = -1

    # shuffle data
    df = df.sample(frac=1).reset_index(drop=True)

    # fetch labels
    if target_colname is not None:
        y = df[target_colname].values
    else:
        raise Exception("Target column name is not provided")
            

    # call stratified kfold class constructor
    kfold = model_selection.StratifiedKFold(n_splits=num_folds)

    # fill in fold no in dataframe
    for foldno, (train_idx, val_idx) in enumerate(kfold.split(X=df, y=y)):
        df.loc[val_idx, "kfolds"] = foldno

    # save to csv
    if save_df:
        df.to_csv("../input/train_folds.csv")
    
    return df
```

```python
df_train_folds = create_folds(df_train, num_folds=5, target_colname="income", save_df=False)
```

### Create LabelEncodings for categorical column using combined train and test df and then split back into train and test

```python
cat_col_names = [c for c in df_train_folds.columns if c not in (numerical_col_names + ["income", "kfolds"])]
cat_col_names
```

```python
def create_labelencoding(df_train, df_test, cat_col_names):
    len_df_train, len_df_test = len(df_train), len(df_test)
    # add a dummy kfolds in df_test since we need to concatenate df_train and df_test to get label encoding on combined
    if "kfolds" not in df_test.columns:
        df_test["kfolds"] = -1
        
    # concatenate df_train and df_test
    df_combined = pd.concat([df_train, df_test], axis=0)
    
    # generate label encoding for every categorical column
    for col in cat_col_names:
        # just for safety convert to str
        df_combined.loc[:,col] = df_combined[col].astype(str)
        # create Label Encoder
        lbl_enc = preprocessing.LabelEncoder()
        # fit the lbl_enc
        lbl_enc.fit(df_combined[col])
        # transform using lbl_enc
        df_combined.loc[:,col] = lbl_enc.transform(df_combined[col])
        
    # revert back to train and test df
    df_train = df_combined[0:len_df_train].reset_index(drop=True)
    
    df_test = df_combined[len_df_train::]
    df_test = df_test.drop("kfolds", axis=1).reset_index(drop=True)
    
    return df_train, df_test
```

### Neural Network Architecture

```python
class MyNetwork(nn.Module):
    def __init__(self, embedd_sizes, num_numerical_features, prob=0.2):
        super().__init__()
        self.embedd_sizes = embedd_sizes
        self.num_numerical_features = num_numerical_features
        
        self.embedding_layer = nn.ModuleList([nn.Embedding(categories, size) for categories,size in self.embedd_sizes])
        n_emb = sum([e.embedding_dim for e in self.embedding_layer]) # combined length of all embedding layers
        self.lin1 = nn.Linear(n_emb + self.num_numerical_features, 256)
        self.lin2 = nn.Linear(256, 128)
        self.lin3 = nn.Linear(128, 64)
        self.lin4 = nn.Linear(64, 32)
        self.lin5 = nn.Linear(32, 1)
        self.bn1 = nn.BatchNorm1d(256)
        self.bn2 = nn.BatchNorm1d(128)
        self.bn3 = nn.BatchNorm1d(64)
        self.bn4 = nn.BatchNorm1d(32)
        self.dropout = nn.Dropout(prob)
        
    def forward(self, x_cat, x_cont):
        # pass the categorical inputs through embedding layer
        x_cat = [e(x_cat[:,i]) for i,e in enumerate(self.embedding_layer)]
        x_cat = torch.cat(x_cat, 1) # dim0 is batch dimension
        x = torch.cat([x_cat, x_cont], axis=1)
        
        x = F.relu(self.bn1(self.lin1(x)))
        x = self.dropout(x)
        x = F.relu(self.bn2(self.lin2(x)))
        x = self.dropout(x)
        x = F.relu(self.bn3(self.lin3(x)))
        x = self.dropout(x)
        x = F.relu(self.bn4(self.lin4(x)))
        x = self.dropout(x)
        x = nn.Sigmoid()(self.lin5(x))
        return x        
```

```python
model = MyNetwork(embedd_sizes, len(numerical_col_names), prob=0.3)
model
```

### set up Dataset, Transforms

```python
class Mydataset(torch.utils.data.Dataset):
    def __init__(self, df, cat_columns, numeric_columns, target_colname, transforms=None):
        self.X_cat = df.loc[:,cat_columns].copy().values.astype(np.int64) # categorical columns
        self.X_num = df.loc[:,numeric_columns].copy().values.astype(np.float32) # numerical columns
        self.y = df.loc[:,target_colname].copy().values.astype(np.float32)
        
        self.transforms = transforms
        
    def __len__(self):
        return len(self.y)
    
    def __getitem__(self, index):
        X_cat, X_num, y = self.X_cat[index], self.X_num[index], self.y[index]
        if self.transforms is not None:
            X_cat, X_num, y  = self.transforms(X_cat, X_num, y)
            
        return X_cat, X_num, y   
```

```python
class ToTensor():
    def __call__(self, X_cat, X_num, y):
        X_cat = torch.tensor(X_cat).type(torch.long)
        X_num = torch.tensor(X_num).type(torch.float)
        y = torch.tensor(y).type(torch.float)

        return X_cat, X_num, y

class Compose():
    def __init__(self, transforms):
        self.transforms = transforms

    def __call__(self, X_cat, X_num, y):
        for t in self.transforms:
            X_cat, X_num, y = t(X_cat, X_num, y)
        return X_cat, X_num, y
```

```python
transform_Compose = Compose([ToTensor()])
```

### set up optimizer, learning rate, device to be trained on, train_one_epoch, val functions

```python
lr=5e-4
# setting up device
device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")

# setting optimizer
params = [p for p in  model.parameters() if p.requires_grad]
optimizer = torch.optim.Adam(params, lr=lr)
#optimizer = torch.optim.SGD(params, lr=lr,
#                            momentum=0.9, weight_decay=0.0005)

# set the model on device
model.to(device)

# criterion
criterion = nn.BCELoss() 
```

```python
def train_one_epoch(model, dataloader, optimizer, device, epoch_num, scheduler=None, fold_no=None):
    # set the model in train mode
    model.train()
    num_batches = len(dataloader)
    batch_num = 0
    loss_list = []
    op_list = []
    tgt_list = []
    for data_cat, data_num, targets in dataloader: # returns batch of data every iteration
        targets = targets.view(targets.size(0),1) # to remove warning
        data_cat = data_cat.to(device)
        data_num = data_num.to(device)
        targets = targets.to(device)
      
        # zero out the gradients so that they do not accumulate (default behaviour) in successive passes
        optimizer.zero_grad()
      
        # forward pass
        op = model(data_cat, data_num)
        # compute loss
        loss = criterion(op, targets) 
        #print(f'loss:{mse_loss}')
        # run backward pass
        loss.backward()
        # update optimizer
        optimizer.step() 
        # step the schedureif enabled
        if scheduler is not None:
            scheduler.step()
       
        batch_num += 1
        if fold_no is None:
            print(f'Epoch:{epoch_num}, Train_batch:{batch_num}/{num_batches}, loss:{loss}')
        else:
            print(f'FoldNo:{fold_no}, Epoch:{epoch_num}, Train_batch:{batch_num}/{num_batches}, loss:{loss}')
            
        loss_list.append(loss.detach().item())
        op_list.extend(op.detach().cpu().numpy())
        tgt_list.extend(targets[:,0].detach().cpu().numpy())
    return loss_list, op_list, tgt_list
```

```python
def eval(model, dataloader, optimizer, device, epoch_num, fold_no=None):
    # set the model in eval mode
    model.eval()
    num_batches = len(dataloader)
    batch_num = 0
    loss_list = []
    op_list = []
    tgt_list = []
    for data_cat, data_num, targets in dataloader: # returns batch of data every iteration
        with torch.no_grad():
            targets = targets.view(targets.size(0),1) # to remove warning
            data_cat = data_cat.to(device)
            data_num = data_num.to(device)
            targets = targets.to(device)
            # forward pass
            op = model(data_cat, data_num)
            # compute loss
            loss = criterion(op, targets) 
             
            batch_num += 1
            if fold_no is None:
                print(f'Epoch:{epoch_num}, Eval_batch:{batch_num}/{num_batches}, loss:{loss.item()}')
            else:
                print(f'FoldNo:{fold_no}, Epoch:{epoch_num}, Eval_batch:{batch_num}/{num_batches}, loss:{loss.item()}')
                
            loss_list.append(loss.item())
            op_list.extend(op.cpu().numpy())
            tgt_list.extend(targets[:,0].cpu().numpy())
    return loss_list, op_list, tgt_list
```

### set up training loop over k folds

```python
num_epochs = 20
bs = 128
num_kfolds = df_train_folds["kfolds"].unique()
stratified_sampler_en = 1 # enables stratified sampling of batches
num_kfolds
```

```python
# code borrowed from https://github.com/ncullen93/torchsample/blob/master/torchsample/samplers.py#L22
class StratifiedSampler(Sampler):
    """Stratified Sampling
    Provides equal representation of target classes in each batch
    """
    def __init__(self, class_vector, batch_size):
        """
        Arguments
        ---------
        class_vector : torch tensor
            a vector of class labels
        batch_size : integer
            batch_size
        """
        self.n_splits = int(class_vector.size(0) / batch_size)
        self.class_vector = class_vector

    def gen_sample_array(self):
        try:
            from sklearn.model_selection import StratifiedShuffleSplit
        except:
            print('Need scikit-learn for this functionality')
        import numpy as np
        
        s = StratifiedShuffleSplit(n_splits=self.n_splits, test_size=0.5)
        X = torch.randn(self.class_vector.size(0),2).numpy()
        y = self.class_vector.numpy()
        s.get_n_splits(X, y)

        train_index, test_index = next(s.split(X, y))
        return np.hstack([train_index, test_index])

    def __iter__(self):
        return iter(self.gen_sample_array())

    def __len__(self):
        return len(self.class_vector)
```

```python
train_stats_folds = []
val_stats_folds = []
best_auc_val = None
for foldno in num_kfolds:
    df_train_ = df_train_folds[df_train_folds.kfolds != foldno].reset_index(drop=True)
    df_val_ = df_train_folds[df_train_folds.kfolds == foldno].reset_index(drop=True)
    
    # construct dataset for the fold
    train_ds = Mydataset(df_train_, cat_col_names, numerical_col_names, "income", transforms=transform_Compose)
    val_ds = Mydataset(df_val_, cat_col_names, numerical_col_names, "income", transforms=transform_Compose)
    
    # construct dataloaders
    if stratified_sampler_en:
        y_train_t = torch.tensor(df_train_.income.values.astype(np.float32)).type(torch.float)
        y_val_t = torch.tensor(df_val_.income.values.astype(np.float32)).type(torch.float)
        stratified_sampler_train = StratifiedSampler(y_train_t, bs)
        stratified_sampler_val = StratifiedSampler(y_val_t, bs)
        train_dl = DataLoader(train_ds, batch_size=bs, shuffle=False, sampler = stratified_sampler_train)
        val_dl = DataLoader(val_ds, batch_size=bs, shuffle=False, sampler = stratified_sampler_val)
    else:
        train_dl = DataLoader(train_ds, batch_size=bs, shuffle=True)
        val_dl = DataLoader(val_ds, batch_size=bs, shuffle=True)
    
    for epoch_num in range(0,num_epochs):
        loss_trainlist, op_trainlist, tgt_trainlist = train_one_epoch(model, train_dl, optimizer, device, epoch_num, \
                                                                 scheduler=None, fold_no=foldno)
        
        loss_vallist, op_vallist, tgt_vallist = eval(model, val_dl, optimizer, device, epoch_num, fold_no=foldno)
        
        
        #compute stats per epoch
        op_train_arr = np.array(op_trainlist)[:,0]
        op_val_arr = np.array(op_vallist)[:,0]
        op_train_prob = op_train_arr.copy()
        op_val_prob = op_val_arr.copy()
        op_train_arr[op_train_arr > 0.5] = 1.0
        op_train_arr[op_train_arr <=0.5] = 0.0
        op_val_arr[op_val_arr > 0.5] = 1.0
        op_val_arr[op_val_arr <= 0.5] = 0.0
        
        train_acc = op_train_arr.mean()
        val_acc = op_val_arr.mean()
        
        avg_trainloss = np.array(loss_trainlist).mean()
        avg_valloss = np.array(loss_vallist).mean()        
        
        
        # AUC on train, val set
        tgt_trainarr = np.array(tgt_trainlist)
        tgt_valarr = np.array(tgt_vallist)
        auc_train = metrics.roc_auc_score(tgt_trainarr, op_train_prob)
        auc_val = metrics.roc_auc_score(tgt_valarr, op_val_prob)
        
        train_stats_folds.append((train_acc, auc_train, avg_trainloss))
        val_stats_folds.append((val_acc, auc_val, avg_valloss))
        
        print(f"fold:{foldno}, epoch:{epoch_num}, train_acc:{train_acc:0.4f}, train_auc:{auc_train:0.4f}, \
                val_acc:{val_acc:.4f}, val_auc:{auc_val:0.4f}, train_loss:{avg_trainloss:.4f}, val_loss:{avg_valloss:0.4f}") 
        
        if best_auc_val==None or auc_val > best_auc_val:
            best_auc_val = auc_val
            torch.save(model.state_dict(), "best_model.pth")
            print("best model saved")
        
```

```python
def get_test_stats(tpr, fpr, op_test_prob, tgt_testlist):
    thresh_opt = thresholds[np.argmax(tpr - fpr)]
    op_test_prob[op_test_prob >= thresh_opt] = 1
    op_test_prob[op_test_prob < thresh_opt] = 0
    prec, recall, f1, support = metrics.precision_recall_fscore_support(np.array(tgt_testlist), op_test_prob)
    confusion_matrix = metrics.confusion_matrix(np.array(tgt_testlist), op_test_prob)
    return prec, recall, f1, support, confusion_matrix
```

### stratified sampling in Pytorch

```python
# code borrowed from https://github.com/ncullen93/torchsample/blob/master/torchsample/samplers.py#L22
class StratifiedSampler(Sampler):
    """Stratified Sampling
    Provides equal representation of target classes in each batch
    """
    def __init__(self, class_vector, batch_size):
        """
        Arguments
        ---------
        class_vector : torch tensor
            a vector of class labels
        batch_size : integer
            batch_size
        """
        self.n_splits = int(class_vector.size(0) / batch_size)
        self.class_vector = class_vector

    def gen_sample_array(self):
        try:
            from sklearn.model_selection import StratifiedShuffleSplit
        except:
            print('Need scikit-learn for this functionality')
        import numpy as np
        
        s = StratifiedShuffleSplit(n_splits=self.n_splits, test_size=0.5)
        X = torch.randn(self.class_vector.size(0),2).numpy()
        y = self.class_vector.numpy()
        s.get_n_splits(X, y)

        train_index, test_index = next(s.split(X, y))
        return np.hstack([train_index, test_index])

    def __iter__(self):
        return iter(self.gen_sample_array())

    def __len__(self):
        return len(self.class_vector)
```
